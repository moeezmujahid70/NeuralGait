```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Assuming 'data' is your dictionary containing the numpy array
ankle_data = data['ankle_data']

# Extract features (all columns except the last one) and target (last column)
features = ankle_data[:, :-1]
targets = ankle_data[:, -1]

# Normalize the features
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(features)
```

```python
print(normalized_features)
```

```python
train_size = int(0.8 * len(normalized_features))
train_features = normalized_features[:train_size]
test_features = normalized_features[train_size:]
train_targets = targets[:train_size]
test_targets = targets[train_size:]
```

```python
print(train_size)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten

# Reshape features for CNN
train_features_cnn = train_features.reshape((train_features.shape[0], train_features.shape[1], 1))
test_features_cnn = test_features.reshape((test_features.shape[0], test_features.shape[1], 1))

# Define CNN model
def create_cnn_model(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(1))  # Output layer for joint moment prediction
    model.compile(optimizer='adam', loss='mse')
    return model

# Create and train CNN model
cnn_model = create_cnn_model(train_features_cnn.shape[1:])
cnn_model.fit(train_features_cnn, train_targets, epochs=100, batch_size=32, validation_split=0.2)

# Predict with CNN
cnn_predictions = cnn_model.predict(test_features_cnn)
```

```python
print(cnn_predictions)
```

```python
import matplotlib.pyplot as plt

# Plot CNN predictions vs actual values
plt.figure(figsize=(14, 7))
plt.plot(test_targets, label='True Values')
plt.plot(cnn_predictions, label='CNN Predictions')
plt.legend()
plt.title('CNN Predictions vs True Values')
plt.show()
```
